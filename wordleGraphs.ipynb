```python
import pandas as pd
from datetime import datetime
```

```python
scores_df = pd.read_csv("myWordleData.csv")
```

```python
scores_df.head()
```

```python
toDate = lambda cell: datetime.strptime(str(cell), '%m/%d/%y')
scores_df["Date"] = scores_df["Date"].apply(toDate)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
# plt.figure(figsize=(30,5))
plot1 = sns.jointplot(x="Date", y="Score", ylim=(0, 7), data=scores_df, hue="Month", height = 10)
# plot1.set_xticklabels(plot.get_xticklabels(), rotation=45)
print()
```
